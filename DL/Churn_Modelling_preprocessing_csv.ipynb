```python
dff.drop(columns=['RowNumber', 'CustomerId', 'Surname'], inplace=True)
```

```python
dff
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
Geography_array = ohe.fit_transform(dff[['Geography']]).toarray()
Geography_array
```

```python
Geography_labels = ohe.categories_
Geography_labels
```

```python
 Geography_flt = pd.DataFrame(Geography_array, columns = Geography_labels)
 Geography_flt
```

```python
Germany = Geography_flt['Germany'] = Geography_flt['Germany'].astype(int)
Germany
```

```python
Spain = Geography_flt['Spain'] = Geography_flt['Spain'].astype(int)
Spain
```

```python
Geography = pd.concat([Germany,Spain], axis=1)
Geography
```

```python
Gender_array = ohe.fit_transform(dff[['Gender']]).toarray()
Gender_array
```

```python
Gender_labels = ohe.categories_
```

```python
Gender_flt = pd.DataFrame(Gender_array, columns= Gender_labels)
Gender_flt
```

```python
Male = Gender_flt['Male'] = Gender_flt['Male'].astype(int)
Male
```

```python
Gender = pd.concat([Male], axis=1)
Gender
```

```python
df_nn = dff.drop(columns=['Geography','Gender'])
df_nn
```

```python
df = pd.concat([df_nn, Germany, Spain, Male], axis=1)
df
```

```python
df.to_csv('Churn_Modelling_preprocessing.csv', index=False)
```
