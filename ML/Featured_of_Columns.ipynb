Basic 7 Questions for Dataset:
1. How Big is the Data ?  Code >>> df.shape
2. How does the Data look like ? Code >>> df.head() or df.sample(5) Note: sample() use for random Data
3. What is the Data Type of Columns ? Code >>> df.info()  
4. Are there any missing values ?  Code >>> df.isnull().sum()
5. How Does The Data Look Mathematically ? Code >>> df.describe()
6. Are there any Duplicate Values ? Code >>> df.duplicated().sum() IF get Any Duplicate Values then use df.drop_duplicates()
7. How is the Corelation Between Columns ? Code >>> df.corr()
