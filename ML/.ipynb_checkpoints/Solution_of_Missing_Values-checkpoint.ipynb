Solution of Missing Values

Source:  https://www.youtube.com/watch?v=uDr67HBIPz8&list=WL&index=342&t=313s

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

Need to Know How many ROW's and Column's in this Dataset

```python
df.shape
```

Show all Null Values

```python
df.isnull()
```

Show Column wise Number of Null Values

```python
df.isnull().sum()
```

Show Total Number of Null Values

```python
df.isnull().sum().sum()
```

Fill Null All Values with specific value

```python
df0 = df.fillna(value = 'BLANK')
df0
```

After Fill all Null values with specific value check the 'sum' of 'isnull'

```python
df0.isnull().sum().sum()
```

Fill Null values with 'previous rows' value

```python
df1 = df.fillna(method = 'pad')
df1
```

Fill Null values with 'Next rows' value

```python
df2 = df.fillna(method = 'bfill')
df2
```

Fill Null values with 'Left side column' value

```python
df3 = df.fillna(method = 'pad', axis = 1)
df3
```

Fill Null values with 'Right side column' value

```python
df4 = df.fillna(method = 'bfill', axis = 1)
df4
```

Fill Null Values with specific column name with specific value

```python
df5 = df.fillna({'who':'me', 'deck':'mango'})
df5
```

Fill Null Value with the mean of a column

```python
df6 = df.fillna(value = df['fare'].mean())    # mean or max or min
df6
```

IF get null value in a Dataset just drop the row of null value

```python
df7 = df.dropna()
df7
```

IF get any null value in a Dataset just drop the row of null value

```python
df8 = df.dropna(how='any')
df8
```

IF get a row that all of the values are null

```python
df9 = df.dropna(how='all')
df9
```

Null Values manage by replace() Function

```python
df10 = df.replace(to_replace = np.nan, value = 'GGG')
df10
```

replace() Function can replace any neumarical value in the Dataset

```python
df11 = df.replace(to_replace = 0.0, value = 5.0)
df11
```
